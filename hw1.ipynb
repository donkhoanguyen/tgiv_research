```python
data = [
    55.3, 55.3, 56.1, 56.1, 57.8, 57.8, 59.8, 59.8, 63.9, 63.9, 65.3, 65.3, 68.7, 68.7,
    73.0, 73.1, 79.3, 79.3, 55.3, 55.9, 55.9, 56.1, 56.1, 56.8, 57.8, 57.9, 57.9, 59.8,
    62.2, 62.2, 64.7, 64.7, 64.7, 65.3, 67.4, 67.4, 69.0, 70.4, 70.4, 73.1, 74.6, 74.6,
    83.0, 83.0, 83.0, 55.9, 55.9, 56.1, 56.1, 56.8, 57.0, 57.0, 57.0, 57.9, 58.8, 58.8,
    58.8, 63.8, 63.8, 63.8, 63.9, 65.1, 65.1, 65.1, 65.3, 67.4, 67.4, 68.7, 68.7, 71.2,
    71.2, 71.2, 73.0, 74.6, 74.6, 79.3, 79.3
]

sorted_data = sorted(data)
```

```python
import numpy as np
from scipy import stats

mean = np.mean(sorted_data)
median = np.median(sorted_data)
mode = stats.mode(sorted_data)
variance = np.var(sorted_data, ddof=1)
std_dev = np.std(sorted_data, ddof=1)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

```python
frequency = {x: sorted_data.count(x) for x in set(sorted_data)}
print("Frequency Distribution:", frequency)
```

```python
data_range = np.max(sorted_data) - np.min(sorted_data)
print(f"Range: {data_range}")
```

```python
Q1 = np.percentile(sorted_data, 25)
Q3 = np.percentile(sorted_data, 75)
IQR = Q3 - Q1

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
```
